```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
    
```

```python
df.columns
```

```python
final_df = df.merge(df_country, on='Country Code', how='left')
```

```python
final_df.head(5)
```

```python
#to check datatype 
final_df.dtypes
```

```python
final_df.columns 
```

```python
country_names=final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#pie chart Top 5 countries that uses zomato
plt.pie(country_val[:5],labels=country_names[:5],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python

ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python

sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
#count plot 
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
#find the countries name that has given 0 rating
final_df[final_df['Rating color']=='White'].groupby(['Aggregate rating','Country']).size().reset_index()
```
